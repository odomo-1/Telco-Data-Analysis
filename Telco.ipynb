## Predicting the Churning of the Customers based on the Factors given

```python
# Importing the cleaned Dataset
Clean_Telco=pd.read_csv('Clean Dataset.csv')
pd.set_option('display.max_columns', None) #show all columns
Clean_Telco.head()
```

```python
#importing the libraries needed
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import StandardScaler

# Import libraries of statsmodels for logistic regression based binary clasification
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels. api as smf
from statsmodels.formula.api import logit
```

```python
# Split data into train and test data
train_data, test_data=train_test_split(Clean_Telco,test_size=0.30)
train_data.head()
```

```python
#Definnig the formula to perform classification based on input features
formula='Churn~gender+SeniorCitizen+Partner+Dependents+tenure+PhoneService+MultipleLines+InternetService+OnlineSecurity+OnlineBackup+DeviceProtection+TechSupport+StreamingTV+StreamingMovies+Contract+PaperlessBilling+PaymentMethod+MonthlyCharges+TotalCharges'
```

```python
#Defining the logistic regression Model
model=logit(formula = formula, data= train_data).fit()
```
